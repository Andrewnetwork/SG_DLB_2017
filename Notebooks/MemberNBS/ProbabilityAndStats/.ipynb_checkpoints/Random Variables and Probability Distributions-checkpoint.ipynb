## Probability Distributions 

The normal distribution is the most common distribution in nature and can be represented by a probability density function. The probability density function of a normal curve can be defined by two parameters: mean and standard deviation. The space under the pdf curve represents the space of all probabilities in the distribution and thus must integrate to 1. If a pdf integrates to less than 1, then the corresponding distribution will have not account for all possible cases. If a pdf integrates to more than 1, then there is more space in our distribution than possible outcomes. In either case, if a pdf does not integrate to 1, it does not accurately represent a probability distribution. 

```python
# Our dense range of inputs to the PDF function for plotting.
x = np.arange(-4,4,0.01)

# The "loc" parameter is the mean. 
# The "scale" parameter is the standard deviation.
z = stat.norm.pdf(x,loc=0,scale=1)
plt.plot(x,z, color="red")

k = stat.norm.pdf(x,loc=0,scale=1.5)
plt.plot(x,k,color="green")

v = stat.norm.pdf(x,loc=1,scale=1)
plt.plot(x,v,color="blue")

plt.show()
print("Area under the normal curve is approximately: {0}.".format(quad(stat.norm.pdf,-np.inf,np.inf)[0]))
```
